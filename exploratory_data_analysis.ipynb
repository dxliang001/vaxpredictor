```python
# imports 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set()
%matplotlib inline
```

```python
df_con = pd.read_csv('data/continents2.csv')
df_con_vac = pd.read_csv('data/country_vaccinations.csv')
df_con_vac_man = pd.read_csv('data/country_vaccinations_by_manufacturer.csv')
df_pop = pd.read_csv('data/population_by_country_2020.csv')
```

```python
df_con.rename(columns={"name": "country"}, inplace=True)
df_con.head()
```

```python
df_con_vac.head()
```

```python
df_con_vac_man.rename(columns={"location": "country"}, inplace=True)
df_con_vac_man.head()
```

```python
df_pop.rename(columns={"Country (or dependency)": "country"}, inplace=True)
df_pop.head()
```

```python
df = pd.merge(df_con_vac, df_pop, on='country')
df.head()
```

```python
df = pd.merge(df, df_con, on='country')
df.head()
```

```python
df = pd.merge(df, df_con_vac_man, on='country', how='left')
df.head()
```

```python
df = df.fillna(0)
df.head()
```

```python
df.to_csv('data\cleaned.csv', index=False)
```
